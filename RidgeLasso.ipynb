```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
X_b = np.c_[np.ones((x.shape[0],1)),x]
X_b
```

```python
a_ridge=1.0
a_lasso=0.1
epochs=1000
lr=0.01
```

```python
def lasso_reg(x,y,alpha,epochs,lr):
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(epochs):
    y_pred=x.dot(theta)
    gradient = (x.T.dot(y_pred-y) + alpha * np.sign(theta))/m
    theta -= lr*gradient
  return theta
```

```python
def ridge_reg(x,y,alpha,epochs,lr):
  m,n=x.shape
  theta=np.zeros(n)
  for _ in range(epochs):
    y_pred=x.dot(theta)
    gradient = (x.T.dot(y_pred-y) + alpha * theta)/m
    theta -= lr*gradient
  return theta
```

```python
t_ridge=ridge_reg(X_b,y,a_ridge,epochs,lr)
t_lasso=lasso_reg(X_b,y,a_lasso,epochs,lr)
```

```python
plt.scatter(x,y,label='Data')
plt.plot(x,X_b.dot(t_ridge),label=f'Ridge(L2) regularization')
plt.plot(x,X_b.dot(t_lasso),label=f'Lasso(L1) regularization')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.show()
```
