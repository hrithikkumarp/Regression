```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
data_ps=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Position_Salaries.csv")
data_ps
```

```python
x=data_ps.iloc[:,1:2].values
x
```

```python
y=data_ps.iloc[:,2:3].values
y
```

```python
from sklearn.linear_model import LinearRegression as LR
lin_reg=LR()
lin_reg.fit(x,y)
```

```python
from sklearn.preprocessing import PolynomialFeatures as PF
poly_reg2=PF(degree=2)
x_poly=poly_reg2.fit_transform(x)

#Linear Regression
lin_reg2=LR()
lin_reg2.fit(x_poly,y)
```

```python
poly_reg3=PF(degree=3)
x_poly3=poly_reg3.fit_transform(x)

lin_reg3=LR()
lin_reg3.fit(x_poly3,y)
```

```python
plt.scatter(x,y,color='r')
plt.plot(x,lin_reg.predict(x),color='r')
# plt.plot(x_poly,lin_reg2.predict(x_poly),color='b')
# plt.plot(x_poly3,lin_reg3.predict(x_poly3),color='g')
plt.title('Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,color='r')
# plt.plot(x,lin_reg.predict(x),color='r')
plt.plot(x,lin_reg2.predict(x_poly),color='b')
# plt.plot(x_poly3,lin_reg3.predict(x_poly3),color='g')
plt.title('Polynomial Linear Regression_Degree2')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
plt.scatter(x,y,color='r')
# plt.plot(x,lin_reg.predict(x),color='r')
# plt.plot(x_poly,lin_reg2.predict(x_poly),color='b')
plt.plot(x,lin_reg3.predict(x_poly3),color='g')
plt.title('Polynomial Linear Regression_Degree3')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
# combined Polynomial Linear Regression
plt.scatter(x,y,color='r')
plt.plot(x,lin_reg.predict(x),color='r')
plt.plot(x,lin_reg2.predict(x_poly),color='b')
plt.plot(x,lin_reg3.predict(x_poly3),color='g')
plt.title('Polynomial Linear Regression')
plt.xlabel('Position level')
plt.ylabel('Salary')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression as LR
from sklearn import metrics
```

```python
x1=real_estate.drop('Y house price of unit area',axis=1)
x1
```

y1=real_estate['Y house price of unit area']
y1

```python
polynomial_reg=PF(degree=2,include_bias=False)
```

```python
poly_feat=polynomial_reg.fit_transform(x1)
poly_feat.shape
```

```python
x_train,y_test = train_test_split(poly_feat,y1,test_size=0.3,random_state=101)
```

```python
polymodel=LR()
polymodel.fit(x_train,y_train)
```

```python
y_pred=polymodel.predict(x_test)
```

```python
pd.DataFrame({'Y_test':y_test,'Y_pred':y_pred,'Residuals':(y_test-y_pred)})
```

```python
poly_MAE = metrics.mean_absolute_error(y_test,y_pred)
poly_MSE = metrics.mean_squared_error(y_test,y_pred)
poly_RMSE = np.sqrt(poly_MSE)
```
